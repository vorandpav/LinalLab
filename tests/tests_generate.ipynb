```python
import numpy as np


def read_matrix(matrix_str) -> np.ndarray:
    lines = matrix_str.strip().split('\n')
    n, m = map(int, lines[0].split())
    matrix = np.array([list(map(float, line.split())) for line in lines[2:]])
    return matrix


m = read_matrix(matrices[4])
centered_m = m - np.mean(m, axis=0)
cov_matrix = np.cov(centered_m, rowvar=False)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
# dict eigenvalue - eigenvectors
eigenvalue_dict = {}
eigenvalues = [float(i) for i in eigenvalues]
eigenvectors = [[float(i) for i in v] for v in eigenvectors]
for i in range(len(eigenvalues)):
    if eigenvalues[i] not in eigenvalue_dict:
        eigenvalue_dict[eigenvalues[i]] = []  # Initialize the list for the eigenvalue
    eigenvalue_dict[eigenvalues[i]].append(eigenvectors[i])
# Sort eigenvalues and eigenvectors
eigenvalues.sort()
eigenvectors = [eigenvalue_dict[i] for i in eigenvalues]


# vector to matrix_string
def matrix_to_string(matrix):
    res = f'"{len(matrix)} {1}\\n"\n"100\\n"\n'
    return res + '\\n"\n"'.join([str(row) for row in matrix]) + '"' + ','


for j in eigenvectors:
    for i in j:
        print(matrix_to_string(i))
```
